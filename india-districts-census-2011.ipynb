```python
census = pd.read_csv('india-districts-census-2011.csv')
census.head()
```

```python
latlong = pd.read_csv('district wise centroids.csv')
latlong.head()
```

```python
col = ['District code', 'District name', 'Population', 'Male', 'Female', 'Literate', 'Male_Literate', 'Female_Literate', 'Hindus', 'Muslims', 'Christians', 'Sikhs', 'Buddhists', 'Jains', 'Housholds_with_Electric_Lighting', 'Households_with_Internet', 'Rural_Households', 'Urban_Households', 'Households', 'Below_Primary_Education', 'Primary_Education', 'Middle_Education', 'Secondary_Education', 'Higher_Education', 'Graduate_Education', 'Having_bathing_facility_Total_Households', 'Having_latrine_facility_within_the_premises_Total_Households']
```

```python
census = census[col]
```

```python
final_df = latlong.merge(census, left_on = 'District', right_on = 'District name').drop(columns = ['District name'])
```

```python
final_df['Sex Ratio'] = round((final_df['Female']/final_df['Male'])*1000)
final_df
```

```python
final_df['Literacy Rate'] = round((final_df['Literate']/final_df['Population'])*100)
final_df
```

```python
final_df.to_csv('final_df.csv', index = False)
```
